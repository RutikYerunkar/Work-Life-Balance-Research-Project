```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('preprocessed_data.csv')
```

```python
data
```

```python
numerical_columns = ['FRUITS_VEGGIES', 'DAILY_STRESS', 'PLACES_VISITED', 'CORE_CIRCLE', 'SUPPORTING_OTHERS', 'SOCIAL_NETWORK', 'ACHIEVEMENT',
                    'DONATION', 'BMI_RANGE', 'TODO_COMPLETED', 'FLOW', 'DAILY_STEPS', 'LIVE_VISION', 'SLEEP_HOURS', 'LOST_VACATION',
                    'DAILY_SHOUTING', 'SUFFICIENT_INCOME', 'PERSONAL_AWARDS', 'TIME_FOR_PASSION', 'WEEKLY_MEDITATION', 'GENDER', 'WORK_LIFE_BALANCE_SCORE',
                    'AGE_21 to 35', 'AGE_36 to 50', 'AGE_51 or more', 'AGE_Less than 20']
scaler = StandardScaler()
data[numerical_columns] = scaler.fit_transform(data[numerical_columns])
```

```python
data
```

```python
# Drop the 'WORK_LIFE_BALANCE_SCORE' column
columns_for_clustering = data.drop(['WORK_LIFE_BALANCE_SCORE','AGE_21 to 35','AGE_36 to 50','AGE_Less than 20','AGE_51 or more'], axis=1)

# Convert the DataFrame to a NumPy array for the clustering algorithm
scaled_data = columns_for_clustering.to_numpy()
```

```python
columns_for_clustering
```

```python
scaled_data
```

```python
# Find the optimal number of clusters (k) using the elbow method
inertia = []
for i in range(2, 10):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.plot(range(2, 10), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# Choose the optimal number of clusters (k) based on the elbow plot
optimal_clusters = 5  # You can change this value according to your elbow plot

# Perform K-Means clustering with the optimal number of clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(scaled_data)

# Assign the cluster labels to the original DataFrame
data['cluster'] = kmeans.labels_
```

```python
data_work_drop = data.drop(['WORK_LIFE_BALANCE_SCORE','AGE_21 to 35','AGE_36 to 50','AGE_Less than 20','AGE_51 or more'], axis=1)
```

```python
data_work_drop
```

```python

# Define the number of clusters you used in KMeans
num_clusters = 5

# Extract the cluster assignments from the DataFrame
clusters = data_work_drop['cluster'].unique()

# Initialize a dictionary to store attributes for each cluster
cluster_attributes = {f'Cluster {cluster}': [] for cluster in clusters}

# Perform KMeans clustering to get the centroids
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(data_work_drop.drop('cluster', axis=1))

# Get the centroids
centroids = kmeans.cluster_centers_

# Analyze the centroids and identify important attributes for each cluster
for cluster_idx in range(num_clusters):
    centroid_values = centroids[cluster_idx]
    sorted_indices = centroid_values.argsort()[::-1]  # Sort indices in descending order
    
    # Set a threshold (you can adjust this as per your requirement)
    threshold = 0.3
    
    # Identify important attributes for this cluster
    important_attributes = data_work_drop.columns[sorted_indices][centroid_values[sorted_indices] >= threshold].tolist()
    
    # Store the important attributes in the dictionary
    cluster_attributes[f'Cluster {cluster_idx}'] = important_attributes

# Print the results
for cluster, attributes in cluster_attributes.items():
    print(f"{cluster} is formed by containing the following attributes:")
    print(attributes)
    print()
```
